##Overloaded term of an index

###Index (noun)

    Relational DB => Database => Tables => Rows      => Columns
    Elasticsearch => Indices  => Types  => Documents => Fields 

###Index (verb)
    To index a document (in Elasticsearch) is to store it in an index for the purpose of searching/retrieval

```python
json_query = {
    "query" : {
        "terms" : {"pres_id" : ["clinton"]}
    },
    "aggregations" : {
        "significantPresWords" : {
            "significant_terms" : { "field" : "transcript" }
        }
    }
}

res = es.search(index="president", doc_type="speeches", body=json_query)
print(json.dumps(res,indent = 2))
```

```python
from elasticsearch import Elasticsearch
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])

terms_json = '{\
  "fields" : ["transcript"],\
  "offsets" : false,\
  "positions" : false,\
  "term_statistics" : true,\
  "field_statistics" : false,\
  "payloads": true \
}'
rs = es.termvector(index='president', doc_type='speeches', id='clinton_speech-4597', body = terms_json)
#print rs['term_vectors']['transcript']['terms']
for term in rs['term_vectors']['transcript']['terms']:
    print term, rs['term_vectors']['transcript']['terms'][term]['term_freq'], \
    rs['term_vectors']['transcript']['terms'][term]['ttf'],\
    rs['term_vectors']['transcript']['terms'][term]['doc_freq']
```
